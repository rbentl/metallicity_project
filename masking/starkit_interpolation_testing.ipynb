```python

grid = load_grid('/u/rbentley/metallicity/grids/bosz_t3500_7000_w20000_24000_R25000.h5')

points, values, wavelengths = grid.index.copy(), grid.fluxes.value.copy(), grid.wavelength.value.copy()

for i in range(len(points)):
    fit = np.polyfit(wavelengths, values[i], 1)
    values[i] = values[i] / (fit[0] * wavelengths + fit[1])
```

```python
points, values, wavelengths = grid.index.copy(), grid.fluxes.value.copy(), grid.wavelength.value.copy()

for i in range(len(points)):
    fit = np.polyfit(wavelengths, values[i], 1)
    values[i] = values[i] / (fit[0] * wavelengths + fit[1])
```

```python
idx = np.where(np.all(points == np.array([3500., 0.0, 0.75, -0.25]), axis=1))[0]

removed_points = points[idx][0]
removed_flux = values[idx][0]

grid.fluxes = np.delete(values, idx, axis=0)
grid.index = np.delete(points, idx, axis=0)
```

```python
interp_test = LinearNDInterpolator(np.delete(points, idx, axis=0), np.delete(values, idx, axis=0))

interp_spec = interp_test(removed_points[0], removed_points[1], removed_points[2], removed_points[3])
```

```python
plt.plot(interp_test(removed_points[0], removed_points[1], removed_points[2], 0.))
```

```python
starspectrum = Spectrum1D.from_array(dispersion=wavelengths, flux=removed_flux,
                                           dispersion_unit=u.angstrom,
                                           uncertainty=removed_flux*(1/100.))
```

```python
interp1 = Interpolate(starspectrum)
#rot1 = RotationalBroadening.from_grid(grid, vrot=np.array([10.0]))
norm1 = Normalize(starspectrum, 2)

model = grid | interp1 | norm1

setattr(model, 'teff_0', removed_points[0])
setattr(model, 'logg_0', removed_points[1])
setattr(model, 'mh_0', removed_points[2])
setattr(model, 'alpha_0', removed_points[3])

w1, f1 = model()

res1 = starspectrum.flux.value - f1
```

```python
data_idx_34 = np.where((w1 >= 22400) & (w1 <= 22750))[0]
data_idx_35 = np.where((w1 >= 21750) & (w1 <= 22100))[0]
data_idx_36 = np.where((w1 >= 21150) & (w1 <= 21500))[0]

cont34 = np.median(starspectrum.flux.value[data_idx_34])
cont35 = np.median(starspectrum.flux.value[data_idx_35])
cont36 = np.median(starspectrum.flux.value[data_idx_36])

print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[22077, 22080], continuum=cont35) #Si line

print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[21377, 21380],continuum=cont36) #Si line
print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[22512, 22515.4], continuum=cont34) #S line

print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[22468.5, 22470.8], continuum=cont34) #Fe line
print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[22478, 22481], continuum=cont34) #Fe line
print rvmeasure.equivalent_width(starspectrum.wavelength.value, starspectrum.flux.value,specRange=[22498.5, 22501.5], continuum=cont34) #Fe line
```

```python
plt.figure(figsize=(16, 12))

plt.plot(wavelengths, removed_flux, label='PHOENIX grid spectrum',
                     color='#000000', linewidth=5.0)

plt.plot(w1, f1, color='#33AFFF', linewidth=5.0)

plt.plot(w1, interp_spec, color='r', linewidth=5.0, label='Scipy interpolator')

plt.plot(w1, res1, label='Starkit interpolated model/residuals',
                    color='#33AFFF', linewidth=5.0)


plt.axhline(y=0.05, color='k', linestyle='--', label='$\pm$ 5%')
plt.axhline(y=-0.05, color='k', linestyle='--')

plt.xlim(21350, 21400)

#plt.ylim(-0.2, 1.6)


plt.legend(loc='center left', fontsize=16)
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Flux')
plt.title('StarKit interpolation of PHOENIX grid point Teff='+str(removed_points[0])+', logg='+str(removed_points[1])+\
          ', [M/H]='+str(removed_points[2])+', [alpha/Fe]='+str(removed_points[3]))
plotlines.oplotlines(angstrom=True, arcturus=True, alpha=0.25, molecules=False, size=15)
```
